```python
from pandas import read_csv

data = read_csv("METxico vs FDA_PubCHEM.csv")
```

```python
from chemplot import Plotter

cp = Plotter.from_smiles(data["Smiles"], target=data["Target"])
```

```python
cp.tsne()
```

```python
plot = cp.visualize_plot(size=15)
plot
```
